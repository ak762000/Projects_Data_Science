```python
X = df.drop(['Attrition'],axis=1)
Y = df['Attrition']
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
```

```python
X = pd.DataFrame(pt.fit_transform(X),columns=X.columns)
X
```

```python
X.skew()
```

```python
#Standard Scaler
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X = pd.DataFrame(scalar.fit_transform(X),columns=X.columns)
X
```

```python
X
```

```python
Y
```

```python
Y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
X,Y = smote.fit_resample(X,Y)
```

```python
Y.value_counts()
```

```python
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras import utils
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=0)  
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Sequential()
model.add(Dense(15,kernel_initializer='he_uniform',activation='relu',input_dim=31))
model.add(Dense(10,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
model.summary()
```

```python
utils.plot_model(model,show_shapes=True)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=10,validation_data=(x_test,y_test))
```

```python
y_pred = model.predict(x_test)
y_pred = y_pred > 0.5
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_lr = lr.predict(x_test)
print(accuracy_score(y_test,y_lr))
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_dt = dt.predict(x_test)
print(accuracy_score(y_test,y_dt))
```

```python
svc = SVC(kernel='rbf')
svc.fit(x_train,y_train)
y_svc = svc.predict(x_test)
print(accuracy_score(y_test,y_svc))
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_rf = rf.predict(x_test)
print(accuracy_score(y_test,y_rf))
```

Machine Learning + Predictive Analysis

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
scaler
```

```python
X = scaler.fit_transform(X)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.30,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

A) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_lr = lr.predict(X_test)
print(y_lr)
```

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import cross_val_score
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_lr))
print("Classification Report : ",classification_report(y_test,y_lr))
print("Cross Val Score : ",cross_val_score(lr,X,Y,cv=5).mean())
```

B) KNNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_knn = knn.predict(X_test)
print(y_knn)
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_knn))
print("Classification Report : ",classification_report(y_test,y_knn))
print("Cross Val Score : ",cross_val_score(knn,X,Y,cv=5).mean())
```

C) Support Vector Machine

```python
from sklearn.svm import SVC
sv = SVC(kernel='rbf')
sv.fit(X_train,y_train)
y_sv = sv.predict(X_test)
print(y_sv)
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_sv))
print("Classification Report : ",classification_report(y_test,y_sv))
print("Cross Val Score : ",cross_val_score(sv,X,Y,cv=5).mean())
```

D) Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_dt = dt.predict(X_test)
print(y_dt)
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_dt))
print("Classification Report : ",classification_report(y_test,y_dt))
print("Cross Val Score : ",cross_val_score(dt,X,Y,cv=5).mean())
```

E) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_rf = rf.predict(X_test)
print(y_rf)
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_rf))
print("Classification Report : ",classification_report(y_test,y_rf))
print("Cross Val Score : ",cross_val_score(rf,X,Y,cv=5).mean())
```
